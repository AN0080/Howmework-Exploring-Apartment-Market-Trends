```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSV data into the variable 'data', specifying ';' as the separator and ',' as the decimal point
try:
    data = pd.read_csv('data.csv', sep=';', decimal=',')
    print("Data loaded successfully.")
except Exception as e:
    print(f"Error loading data: {e}")
    exit()  # Terminate execution to avoid further errors

# Ensure the dataset is not empty
if data.empty:
    print("Error: The dataset is empty.")
    exit()

# Handle missing values to avoid issues in visualization
data = data.dropna()
```

```python
# Visualization 1: Bar plot of apartment prices by neighborhood
plt.figure(figsize=(10,6))
sns.barplot(data=data, x='neighborhood', y='apt_price', hue='neighborhood', palette='viridis', legend=False)
plt.title('Apartment Prices by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Apartment Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 2: Line plot of apartment price vs. square meter
plt.figure(figsize=(10,6))
sns.lineplot(data=data, x='square meter', y='apt_price', marker='o')
plt.title('Apartment Price vs. Square Meter')
plt.xlabel('Square Meter')
plt.ylabel('Apartment Price')
plt.grid(True)
plt.show()
```

```python
# Visualization 3: Scatter plot of floor vs. apartment price
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='floor', y='apt_price', alpha=0.7)  # Adjust transparency
plt.title('Floor vs. Apartment Price')
plt.xlabel('Floor')
plt.ylabel('Apartment Price')
plt.grid(True)
plt.show()
```

```python
# Visualization 4: Heatmap displaying correlations between numeric variables
plt.figure(figsize=(10,6))
numeric_data = data.select_dtypes(include=['number'])  # Select only numerical columns
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Data')
plt.show()
```

```python
# Visualization 5: Histogram of apartment price distribution
plt.figure(figsize=(10,6))
sns.histplot(data=data, x='apt_price', bins=20, kde=True, color='blue')
plt.title('Distribution of Apartment Prices')
plt.xlabel('Apartment Price')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Visualization 6: Box plot of apartment prices by furniture condition
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='furniture', y='apt_price', hue='furniture', palette='coolwarm', legend=False)  
plt.title('Apartment Prices by Furniture Condition')
plt.xlabel('Furniture Condition')
plt.ylabel('Apartment Price')
plt.xticks(rotation=45)
plt.show()
```
